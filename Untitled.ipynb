```python
df = pd.DataFrame({'user_message': messages, 'message_date': dates})

# Remove trailing ' - '
df['message_date'] = df['message_date'].str.strip(' - ')

# Define a function to parse dates manually and log issues
def parse_date(date_str):
    try:
        return pd.to_datetime(date_str, format='%m/%d/%y, %H:%M')
    except ValueError:
        print(f"Failed to parse date: {date_str}")
        return pd.NaT

# Apply the parsing function
df['message_date'] = df['message_date'].apply(parse_date)

# Rename the column
df.rename(columns={'message_date': 'date'}, inplace=True)

# Display the first few rows of the DataFrame
df.head()
```

```python
df.shape
```

```python
users = []
messages = []
for message in df['user_message']:
    entry = re.split('([\w\W]+?):\s', message)
    if entry[1:]:  # user name
        users.append(entry[1])
        messages.append(" ".join(entry[2:]))
    else:
        users.append('group_notification')
        messages.append(entry[0])

df['user'] = users
df['message'] = messages
df.drop(columns=['user_message'], inplace=True)

df.head()
```

```python
df['date'].dt.year
```

```python
df['only_date'] = df['date'].dt.date
df['year'] = df['date'].dt.year
df['month_num'] = df['date'].dt.month
df['month'] = df['date'].dt.month_name()
df['day'] = df['date'].dt.day
df['day_name'] = df['date'].dt.day_name()
df['hour'] = df['date'].dt.hour
df['minute'] = df['date'].dt.minute
```

```python

import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import poisson

# Parameters
lambdas = [1, 3, 5]  # Different lambda values to plot
x = np.arange(0, 15)  # Range of x values (number of events)

# Plotting
plt.figure(figsize=(12, 8))

for lam in lambdas:
    pmf = poisson.pmf(x, lam)
    plt.plot(x, pmf, 'o', label=f'Î» = {lam}')

plt.title('Poisson Distribution PMF for Different Î»')
plt.xlabel('Number of Events (k)')
plt.ylabel('Probability')
plt.legend()
plt.grid(True)
plt.show()
```
